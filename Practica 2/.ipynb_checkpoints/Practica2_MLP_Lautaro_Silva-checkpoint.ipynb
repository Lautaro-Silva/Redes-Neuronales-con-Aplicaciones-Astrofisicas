## Descripci√≥n del dataset:

**green_color**: cantidad de luz medida en la banda de frecuencia que los astrof√≠sicos llaman verde (longitud de onda entre 464 y 528~nm).    

**red_color**: flujo de luz medida en la banda de frecuencia rojo (longitud de onda entre 658 y 796~nm)

**infrared_color**: flujo de luz medida en la banda de frecuencia infrarroja (longitud de onda entre 806 y 955~nm)

**de_vaucouleurs_fit_axial_ratio**: radio axial asociado al ajuste de De Vaucouleurs (https://es.wikipedia.org/wiki/Ley_de_De_Vaucouleurs) a la imagen bidimensional de la galaxia

**exponential_fit_axial_ratio**: radio axial asociado a un ajuste exponencial a la imagen bidimensional de la galaxia

**exponential_disc_fit_log_likelihood**: log-verosimilitud asociada al ajuste exponencial a la imagen bidimensional de la galaxia

**de_vaucouleurs_fit_log_likelihood**: log-verosimilitud asociada al ajuste de De Vaucouleurs a la imagen bidimensional de la galaxia

**star_log_likelihood**: log-verosimilitud de que el objeto sea una fuente puntual (estrella) en vez de una galaxia

**number_of_votes**: n√∫mero total de votos de los voluntarios

**probability_elliptical**: fracci√≥n de votos para morfolog√≠a el√≠ptica

**probability_spiral**: fracci√≥n de votos para morfolog√≠a espiral

**probability_unknown**: fracci√≥n de votos para morfolog√≠a desconocida

**probability_merger**: fracci√≥n de votos para morfolog√≠a de fusi√≥n

**class**: morfolog√≠a con la mayor fracci√≥n de votos

```python
# Ejemplo para Remover outliers en altura de personas


# DataFrame ficticio con alturas (en cm) y un par de valores extremos (outliers)
df_personas = pd.DataFrame({
    'altura_cm': np.concatenate([
        np.random.normal(170, 10, 100),  # la mayor√≠a de las personas
        [120, 220]  # dos outliers
    ])
})

# Calculamos media y desv√≠o est√°ndar
media = df_personas['altura_cm'].mean()
std = df_personas['altura_cm'].std()

# Definimos l√≠mites (por ejemplo, 3 sigmas)
lim_inf, lim_sup = media - 3*std, media + 3*std

# Filtramos el DataFrame para sacar los outliers
df_personas_limpio = df_personas[
    (df_personas['altura_cm'] >= lim_inf) &
    (df_personas['altura_cm'] <= lim_sup)
]

print(f"Cantidad antes de limpiar: {len(df_personas)}")
print(f"Cantidad despu√©s de limpiar: {len(df_personas_limpio)}")

# Opcional: Visualizamos antes y despu√©s
import matplotlib.pyplot as plt

plt.subplot(1,2,1)
plt.hist(df_personas['altura_cm'], bins=15, color='skyblue', edgecolor='k')
plt.title('Con outliers')
plt.xlabel('Altura (cm)')
plt.ylabel('Cantidad')

plt.subplot(1,2,2)
plt.hist(df_personas_limpio['altura_cm'], bins=15, color='lime', edgecolor='k')
plt.title('Sin outliers')
plt.xlabel('Altura (cm)')

plt.tight_layout()
plt.show()
```

La **correlaci√≥n de Pearson** entre dos variables $ X $ e $ Y $ se define como:


$$r_{XY} = \frac{ \sum_{i=1}^{n} (X_i - \overline{X})(Y_i - \overline{Y}) }
{ \sqrt{ \sum_{i=1}^{n} (X_i - \overline{X})^2 } \; \sqrt{ \sum_{i=1}^{n} (Y_i - \overline{Y})^2 } }$$

donde:
- $ X_i, Y_i $: valores individuales de cada variable
- $ \overline{X}, \overline{Y} $: promedio de cada variable
- $ n $: cantidad de datos

Valores cercanos a 1 o -1, indican alta correlacion, y valores cercanos a 0, no hay relaci√≥n lineal.
